```python
def getM(k = 2, printmode=True):
  M = np.zeros((2**k, 2**k))

  for i in range(2**k):
    si = np.binary_repr(i, width=k)
    zi = np.array([int(x) for x in si])
    if printmode:
      print (si)
    for j in range(2**k):
      sj = np.binary_repr(j, width=k)
      zj = np.array([int(x) for x in sj])
      M[i, j] = sum(zi != zj)
  return M

getM(1)
```

```python
def getHfromM(M):
  n = M.shape[0]
  H = []
  for i in range(n-1):
    for j in range(i+1, n):
      print (i, j)
      z = M[i,:]!=M[j,:]
      if ~z.all():
        H.append(z)
  return(np.array(H)+0)

H = getHfromM(M)
print (H)
```
